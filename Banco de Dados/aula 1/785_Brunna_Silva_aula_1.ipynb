```python
import psycopg2
import pandas as pd
```

```python
conn = psycopg2.connect(
    host="localhost",
    database="northwind",
    user="postgres",
    password="1234")

cur = conn.cursor()

#Pandas Table
def create_pandas_table(sql_query, database = conn):
    table = pd.read_sql_query(sql_query, database)
    return table
```

```python
#Testando conexão
customers_info = create_pandas_table("SELECT * FROM customers LIMIT 10")
customers_info
```

```python
#1. Selecione os id's únicos de territórios da tabela employee_territories.
employees_info = create_pandas_table("SELECT DISTINCT territory_id \
                                    FROM EMPLOYEE_TERRITORIES LIMIT 10")
employees_info
```

```python
#2.A. Todos os nomes dos empregados;
employees_info = create_pandas_table("SELECT first_name AS nome, last_name AS sobrenome \
                                     FROM employees")
employees_info
```

```python
#2.B. Os sobrenomes distintos dos empregrados;
employees_info = create_pandas_table("SELECT DISTINCT last_name AS sobrenome \
                                     FROM employees")
employees_info
```

```python
#2.C. Empregados que nasceram após 1960;
employees_info = create_pandas_table("SELECT first_name AS nome, last_name AS sobrenome,\
                                    birth_date FROM employees \
                                    WHERE date_part('year', birth_date) > 1960")
employees_info
```

```python
#2.D. Empregados que foram contratados em 1993;
employees_info = create_pandas_table("SELECT first_name AS nome, last_name AS sobrenome,\
                                    hire_date FROM employees \
                                    WHERE date_part('year', hire_date) = 1993")
employees_info
```

```python
#2.E. Empregados que nasceram entre 1950 e 1958 (inclusos);
employees_info = create_pandas_table("SELECT first_name AS nome, last_name AS sobrenome,\
                                    birth_date FROM employees \
                                    WHERE date_part('year', birth_date) \
                                    BETWEEN 1950 AND 1958")
employees_info
```

```python
#2.F. Empregados que nasceram em 1963 e foram contratados em 1993;
employees_info = create_pandas_table("SELECT first_name AS nome, last_name AS sobrenome,\
                                    birth_date FROM employees \
                                    WHERE date_part('year', birth_date) = 1963 AND \
                                    date_part('year', hire_date) = 1993")
employees_info
```

```python
#2.G. Todos os empregrados que reportam o chefe de id 5;
employees_info = create_pandas_table("SELECT first_name AS nome, last_name AS sobrenome, \
                                    reports_to FROM employees WHERE reports_to = 5")
employees_info
```

```python
#2.H. Todos os empregados que moram em Seattle ou Kirkland.
employees_info = create_pandas_table("SELECT first_name AS nome, last_name AS sobrenome \
                                    FROM employees WHERE city IN ('Seattle','Kirkland')")
employees_info
```

```python
#3. Selecione da tabela order_details:
order_info = create_pandas_table("SELECT * FROM order_details")
order_info
```

```python
#3.A. Produtos (product_id) com mais de 50 unidades vendidas;
order_info = create_pandas_table("SELECT product_id, quantity FROM order_details \
                                WHERE quantity > 50")
order_info
```

```python
#3.B. Produtos com mais de 0.2 de desconto;
order_info = create_pandas_table("SELECT product_id, discount FROM order_details \
                                WHERE discount > 0.21") #hihihi, gambiarra!
order_info
```

```python
#3.B. Produtos com 0.05 ou menos de desconto;
order_info = create_pandas_table("SELECT product_id, discount FROM order_details \
                                WHERE discount <= 0.5") 
order_info
```

```python
#4. Selecione da tabela orders:
orders_info = create_pandas_table("SELECT * FROM orders LIMIT 5") 
orders_info
```

```python
#4.A. Ordens realizadas antes Setembro de 1996:
orders_info = create_pandas_table("SELECT order_id, order_date FROM orders \
                                WHERE date_part('year', order_date) <= 1996 \
                                AND date_part('month', order_date) <= 09") 
orders_info
```

```python
#4.B. Ordens enviadas em Setembro de 1996.
orders_info = create_pandas_table("SELECT order_id, shipped_date FROM orders \
                                WHERE date_part('year', shipped_date) = 1996 \
                                AND date_part('month', shipped_date) = 09 LIMIT 5") 
orders_info
```

```python
#4.C. Ordens que possuam o campo região preenchido.
orders_info = create_pandas_table("SELECT order_id, ship_region AS region FROM orders \
                                WHERE ship_region IS NOT NULL LIMIT 5") 
orders_info
```

```python
#4.D. As primeiras 5 linhas da tabela, 
#reescrevendo o nome das colunas de data em português.
orders_info = create_pandas_table("SELECT order_id, customer_id, employee_id, \
                                order_date AS data_pedido, required_date AS data_requerimento,\
                                shipped_date AS data_envio, ship_via, freight, ship_name,\
                                ship_address, ship_city, ship_region, ship_postal_code,\
                                ship_country FROM orders LIMIT 5") 
orders_info
```

```python
#5. Selecione da tabela suppliers:
suppliers_info = create_pandas_table("SELECT * FROM suppliers LIMIT 5") 
suppliers_info
```

```python
#5.A. Todos os contatos cujo nome comece com a letra M;
suppliers_info = create_pandas_table("SELECT contact_name FROM suppliers\
                                    WHERE contact_name LIKE 'M%' LIMIT 5") 
suppliers_info
```

```python
#5.B. Todos os contatos que trabalhem com vendas e morem nos países nórdicos.
suppliers_info = create_pandas_table("SELECT contact_name, country FROM suppliers\
                                    WHERE country in ('Denmark', 'Sweden', 'Norway', 'Finland')") 
suppliers_info
```
